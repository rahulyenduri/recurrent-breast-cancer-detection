```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
```

```python
x=data
x.head()
```

# We will start preprocessing the data 
## The dataset consists of various string values that are not understood by all ML models
**Hence we replace the various strings in various columns with their proper integer values**

```python
x['menopause'] = x['menopause'].replace(['premeno'],0)
x['menopause'] = x['menopause'].replace(['ge40'],1)
x['menopause'] = x['menopause'].replace(['lt40'],2)
x.head()
```

```python
x['age'] = x['age'].replace(['20-29'],0)
x['age'] = x['age'].replace(['30-39'],1)
x['age'] = x['age'].replace(['40-49'],2)
x['age'] = x['age'].replace(['50-59'],3)
x['age'] = x['age'].replace(['60-69'],4)
x['age'] = x['age'].replace(['70-79'],5)
x.head()
```

```python
x['tumor-size'] = x['tumor-size'].replace(['0-4'],0)
x['tumor-size'] = x['tumor-size'].replace(['5-9'],1)
x['tumor-size'] = x['tumor-size'].replace(['10-14'],2)
x['tumor-size'] = x['tumor-size'].replace(['15-19'],3)
x['tumor-size'] = x['tumor-size'].replace(['20-24'],4)
x['tumor-size'] = x['tumor-size'].replace(['25-29'],5)
x['tumor-size'] = x['tumor-size'].replace(['30-34'],6)
x['tumor-size'] = x['tumor-size'].replace(['35-39'],7)
x['tumor-size'] = x['tumor-size'].replace(['40-44'],8)
x['tumor-size'] = x['tumor-size'].replace(['45-49'],9)
x['tumor-size'] = x['tumor-size'].replace(['50-54'],10)
x.head()
```

```python
x['inv-nodes'] = x['inv-nodes'].replace(['0-2'],0)
x['inv-nodes'] = x['inv-nodes'].replace(['3-5'],1)
x['inv-nodes'] = x['inv-nodes'].replace(['6-8'],2)
x['inv-nodes'] = x['inv-nodes'].replace(['9-11'],3)
x['inv-nodes'] = x['inv-nodes'].replace(['12-14'],4)
x['inv-nodes'] = x['inv-nodes'].replace(['15-17'],5)
x['inv-nodes'] = x['inv-nodes'].replace(['24-26'],6)
x.head()
```

```python
x['node-caps'] = x['node-caps'].replace(['no'],0)
x['node-caps'] = x['node-caps'].replace(['yes'],1)
x.head()
```

```python
x['breast'] = x['breast'].replace(['left'],0)
x['breast'] = x['breast'].replace(['right'],1)
x.head()
```

```python
x['breast-quad'] = x['breast-quad'].replace(['central'],0)
x['breast-quad'] = x['breast-quad'].replace(['left_low'],1)
x['breast-quad'] = x['breast-quad'].replace(['left_up'],2)
x['breast-quad'] = x['breast-quad'].replace(['right_low'],3)
x['breast-quad'] = x['breast-quad'].replace(['right_up'],4)
x.head()
```

```python
x['irradiat'] = x['irradiat'].replace(['no'],0)
x['irradiat'] = x['irradiat'].replace(['yes'],1)
x.head()
```

```python
x['Class'] = x['Class'].replace(['Non-Recurrence Events'],0.0)
x['Class'] = x['Class'].replace(['Recurrence Events'],1.0)
x.head()
```

## We clean the data by filling the missing and NaN values with the median of the data

```python
x['node-caps'] = x['node-caps'].fillna(x['node-caps'].median())
x['breast-quad'] = x['breast-quad'].fillna(x['breast-quad'].median())
x.head()
```

### We remove the id and Class columns before training dataset because:
#### 1. id does not help us finding the target value it is just for indexing
#### 2. Class is the target value hence removing it from the training dataset

```python
y=data.Class
l=['id','Class']
x=data.drop(l,axis=1)
x.head()
```

### Now we split the dataset into training and testing datasets

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=5) 
```

### Here we preprocess the data again to scale values between 0 and 1 because few ML models work efficiently if the data is scaled

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

## KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_sc,y_train)
y_pred=knn.predict(x_test_sc)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
bar_data= {"Voting Classifier":0.9635549872122762,"Adaboost Classifier": 0.9079159935379645,"Multi Layer Perceptron":0.9201277955271565,"Histogram based GBM":0.9635549872122762}
models=list(bar_data.keys())
scores=list(bar_data.values())
fig = plt.figure(figsize = (8, 8))
plt.ylim(0.86, 0.98,0.01)
plt.bar(models[0],scores[0], color ='chartreuse',width = 0.3)
plt.bar(models[1],scores[1],color ='orange',width = 0.3)
plt.bar(models[2],scores[2], color ='crimson',width = 0.3)
plt.bar(models[3],scores[3], color ='teal',width = 0.3)
plt.xlabel("Models used")
plt.ylabel('Cohen_Kappa')
plt.title("Comparision between Cohen_Kappa values of various models")
for index, value in enumerate(scores):
    plt.text(index-0.11, value+0.001,(str(value))[:5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
barWidth = 0.1
pred1 = [0.9782,0.962,0.969,0.9782]
pred2 = [0.9767,0.913,0.943,0.913]
pred3 = [0.96,0.923,0.957,0.923]
pred4 = [0.9687,0.947,0.959,0.941]
br1 = np.arange(len(pred1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
fig = plt.figure(figsize = (10, 5))
plt.ylim(0.9, 1,0.1)
plt.bar(br1,pred1, color ='r', width = barWidth,edgecolor ='black', label='Voting Classifier')
plt.bar(br2,pred2, color ='g', width = barWidth,edgecolor ='black', label='Adaboost')
plt.bar(br3,pred3, color ='b', width = barWidth,edgecolor ='black', label='Multi-Layer Perceptron')
plt.bar(br4,pred4, color ='lime', width = barWidth,edgecolor ='black', label='Histogram Based')
plt.xlabel('Evaluation Metrics', fontweight ='bold', fontsize = 15)
plt.ylabel('Values', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(pred1))],
        ["Precision","Recall","F1-Score","Specificity"])
 
plt.legend()
plt.show()
```
